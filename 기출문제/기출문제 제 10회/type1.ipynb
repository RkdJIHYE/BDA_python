```python
df = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/data/main/exam/10_1_1.csv")
```

```python
print(df.head())
print(df.info())
```

```python
#1번문제 : 3째로 높은 정답률을 구하시오 
one_num = df.groupby(["소주제"])['정답여부'].sum() #1만 합해짐
number = df.groupby(['소주제'])['정답여부'].count() #전체 개수 
print(one_num)
#print(number)

ratio = np.round(one_num/number,2)
print(ratio.sort_values(ascending=False))

#정답은 경제 
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/data/main/exam/10_1_2.csv")
print(df.info())
```

```python
print(df.head())
```

```python
#date 를 연도 월 로 분리 / 연도 월별 price 의 합계 / 두번째 큰 매출액 구하기 

df['date']=pd.to_datetime(df['date'])
```

```python
df.info()
#오브젝트 -> date로 형식이 변한것을 확인 가능 
```

```python
df['year']=df['date'].dt.year
df['month']=df['date'].dt.month
```

```python
print(df)
```

```python
month_sale = df.groupby(["year","month"])["price"].sum().sort_values(ascending=False)
print(month_sale)
```

```python
category_sale = df.groupby(["year","month","category"])["price"].sum().sort_values(ascending=False).reset_index()
```

```python
#print(category_sale)
df=category_sale.loc[(category_sale['year']==2024)&(category_sale['month']==12),:]
print(df)
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/data/main/exam/10_1_3.csv")
print(df.info())
print(df.head())
```

```python
df['count']=df['message'].str.split(' ').apply(len)
```

```python
mean_cnt = df.groupby(['label'])['count'].mean()
print(mean_cnt)
```

```python
abs_num = np.round(abs(mean_cnt['ham']-mean_cnt['spam']),3)
print(abs_num)
```
