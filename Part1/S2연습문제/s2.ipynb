```python
#리스트를 -> 튜플 -> 집합 개별 요소의 개수는?
nums=[5,10,10,15]
print(nums)
print(len(nums))

tnums = tuple(nums)
print(tnums) 
print(len(tnums))

#set() 집합
snums = set(tnums)

print(snums)
print(len(snums))
```

```python
#다음 딕셔너리에서 age 를 28로 수정하고 city 키-값 쌍을 삭제한 후 딕셔너리를 출력하시오 

profile={
    "name":"Jane",
    "age":27,
    "city":"Busan"
}



profile["age"]=28
del profile["city"]

print(profile)
```

```python
#다음 두 집합의 합,교,차집합을 각각 출력하시오
set_x = {1,2,3,4}
set_y = {3,4,5,6}


sxy =set_x.union(set_y)
print(sxy)

ixy = set_x.intersection(set_y)
print(ixy)

dxy = set_x.difference(set_y)
print(dxy)
```
