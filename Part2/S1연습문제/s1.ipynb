```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/data/main/grade.csv")
```

```python
#df 데이터 프레임의 정보 출력, 각 칼럼의 데이터 타입 확인

print(df.info())
```

```python
#midterm 점수가 85점 이상인 학생의 데이터
print(df.loc[df['midterm']>=85])
```

```python
#final 점수를 기준으로 데이터 프레임을 내림차순 정렬, 정렬된 데이터 프레임의 첫 5행을 출력

rdf = df.sort_values(by='final',ascending=False)
print(rdf.head())
```

```python
#gender 칼럼을 기준으로 데이터 프레임을 그룹화하고, 각 그룹별 Midterm 과 final 의 평균을 계산하여 출력하시오

gdf = df.groupby('gender')[['midterm','final']].mean()
print(gdf.head())
```

```python
#student_id 칼럼을 문자열 타입으로 변환하고 변환된 데이터 프레임의 정보를 출력하시오.

df = df.astype({'student_id':'str'})
print(df.info())
```

```python
#assignment 점수의 최댓값과 최솟값을 가지는 행을 각각 출력하시오.

max_idx = df['assignment'].idxmax()
min_idx = df['assignment'].idxmin()


print(df.loc[max_idx])
print(df.loc[min_idx])
```

```python
#midterm,final,assignment 점수의 평균을 계산하여 average 칼럼을 추가하고, 첫 5행 출력하기 

df['average']=df[['midterm','final','assignment']].mean(axis=1)
print(df.head())
```

```python
#데이터 프레임에 결측치가 있는지 확인하고, 결측치를 포함한 행을 제거후 데이터 프레임의 정보를 출력하시오

print(df.isna().sum())
```

```python
clean_df=df.dropna()
print(clean_df.info())
```

```python
#아래의 추가 데이터 생성, 기존 데이터 프레임과 student_id를 기준으로 병합하여 출력

additional_data={
    'student_id' : ['1','3','5','7','9'],
    'club' : ['Art','Science','Math','Music','Drama']
}

df_additional = pd.DataFrame(additional_data)
```

```python
merged_df = pd.merge(df,df_additional,on='student_id',how='left')
print(merged_df)
```

```python
#gender 인덱스로, studnet_id를 열로 사용하여 average 점수에 대한 피벗 테이블 생성 및 출력

pivot_df = df.pivot_table(values='average',index='gender',columns='student_id')
print(pivot_df)
```

```python
#midterm,final,assignment의 평균을 구하고 ,average 열을 생성

df['average']=df[['midterm','final','assignment']].mean(axis=1)

melted_df = pd.melt(df,id_vars=['student_id','name','gender'],value_vars=['midterm','final','assignment','average'],
                    var_name='variable',value_name='score')
group_df = melted_df.groupby(['gender','variable'])['score'].mean().reset_index()

print(group_df)
```

```python
#최대 평균 성적을 가진 학생의 이름과 평균 성적 출력

max_idx = df['average'].idxmax()
max_avg_st = df.loc[max_idx,['name','average']]

print(f"최대 평균성적을 가진 학생의 이름은 {max_avg_st['name']}이고 점수는 {max_avg_st['average']}입니다.")
```
