```python
import pandas as pd
```

```python
data = {
    'Date' : ['2024-07-01','2024-07-02','2024-07-03','2024-07-03'],
    'Temperature' : [10,20,25,20],
    'Humidity' : [60,65,70,21]
    
}
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
df_melted = pd.melt(df,
                    id_vars=['Date'],
                    value_vars=['Temperature','Humidity'],
                    var_name='Variable',
                    value_name = 'Value'
                    )
```

```python
print(df_melted)
```

```python
#다시 원상 복구 하기 위해서는 pivot 을 이용해 준다. 그러나 고유한 조합이어야 한다.

df_melted2=pd.melt(df.reset_index(),
                    id_vars=['index'],
                    value_vars=['Temperature','Humidity'],
                    var_name='Variable',
                    value_name = 'Value'
                    )

print(df_melted2)
```

```python
df_pivoted = df_melted2.pivot(index = 'index',
                        columns = 'Variable',
                        values = 'Value')

print(df_pivoted)
```

```python
#위에서는 고유한 Index 를 추가해 주거나 아니면 -> pivot 대신 pivot_table()을 이용해주면 중복있는 날짜의 값을 평균으로 대치해 준다 !

df_pivot_table = df_melted.pivot_table(index = 'Date',
                                       columns = 'Variable',
                                       values= 'Value').reset_index()
```

```python
print(df_pivot_table)
```
